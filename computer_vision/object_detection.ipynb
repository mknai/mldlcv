```python
from imageai.Detection import ObjectDetection
import os
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yolo_path = "/Users/noor-4785/Downloads/yolo.h5"

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(yolo_path)
detector.loadModel()
```

```python
image_path = "/Users/noor-4785/Downloads/11603_car.jpg"
output_path = "/Users/noor-4785/Downloads/11603_car_yolo.jpg"
detections = detector.detectObjectsFromImage(input_image=image_path, output_image_path=output_path, minimum_percentage_probability=30)
```

```python
image = cv2.imread(output_path)
plt.imshow(image)
plt.show()
```

```python
for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```
